```python

# --- Convert to Million ---
seasons = (
    df.groupby("Season Number")['Original Ask Amount']
      .sum()
      .sort_index() / 10_00_000
)

deal_amount = (
    df.groupby("Season Number")['Total Deal Amount']
      .sum()
      .sort_index() / 10_00_000
)

invested_amount = (
    df.groupby("Season Number")['Investment Amount Per Shark']
      .sum()
      .sort_index() / 10_00_000
)

season_numbers = seasons.index

# --- Colors ---
colors = plt.cm.tab20.colors[:len(season_numbers)]
color_map = dict(zip(season_numbers, colors))

# --- Figure ---
fig, ax = plt.subplots(1, 3, figsize=(18, 10))
fig.suptitle(
    "Original Ask vs Total Deal vs Invested Amount ($ in Million)",
    fontsize=14,y=0.98
)

# --- Plot 1 ---
bars0 = ax[0].barh(
    season_numbers,
    seasons.values,
    color=[color_map[s] for s in season_numbers]
)
ax[0].set_xlabel("Original Ask Amount ($ Million)")
ax[0].set_ylabel("Season Number")
ax[0].set_title("Original Ask")
ax[0].bar_label(bars0, fmt="%.1fM", padding=3)

# --- Plot 2 ---
bars1 = ax[1].barh(
    season_numbers,
    deal_amount.values,
    color=[color_map[s] for s in season_numbers]
)
ax[1].set_xlabel("Total Deal Amount ($ Million)")
ax[1].set_title("Total Deal Amount")
ax[1].bar_label(bars1, fmt="%.1fM", padding=3)

# --- Plot 3 ---
bars2 = ax[2].barh(
    season_numbers,
    invested_amount.values,
    color=[color_map[s] for s in season_numbers]
)
ax[2].set_xlabel("Invested Amount by Sharks ($ Million)")
ax[2].set_title("Invested Amount")
ax[2].bar_label(bars2, fmt="%.1fM", padding=3)

# --- Legend ---
legend_patches = [
    Patch(color=color_map[s], label=f"Season {s}")
    for s in season_numbers
]

fig.legend(
    handles=legend_patches,
    title="Seasons",
    loc="upper center",
    ncol=6,
    bbox_to_anchor=(0.5, 0.94)
)

plt.tight_layout(rect=[0, 0, 1, 0.88])

plt.show()
```

```python

# --- Highest ask per season ---
idx_max = df.groupby("Season Number")["Original Ask Amount"].idxmax()
result_max = df.loc[
    idx_max, ["Season Number", "Startup Name", "Original Ask Amount"]
].sort_values("Season Number")

# --- Lowest ask per season ---
idx_min = df.groupby("Season Number")["Original Ask Amount"].idxmin()
result_min = df.loc[
    idx_min, ["Season Number", "Startup Name", "Original Ask Amount"]
].sort_values("Season Number")

# --- Seasons ---
season_numbers = result_max["Season Number"]

# --- One color per season ---
colors = plt.cm.tab20.colors[:len(season_numbers)]
color_map = dict(zip(season_numbers, colors))

# --- Plot ---
fig, ax = plt.subplots(1, 2, figsize=(18, 10))
fig.suptitle(
    "Highest vs Lowest Original Ask per Season ($ Million)",
    fontsize=16,
    fontweight="bold",
    y=0.98
)


# --- Highest ---
bars0 = ax[0].barh(
    result_max["Startup Name"],
    result_max["Original Ask Amount"] / 10_00_000,
    color=[color_map[s] for s in result_max["Season Number"]]
)
ax[0].set_title("Highest Original Ask")
ax[0].set_xlabel("Amount ($ Million)")
ax[0].bar_label(bars0, fmt="$%.2fM", padding=3)

# --- Lowest ---
bars1 = ax[1].barh(
    result_min["Startup Name"],
    result_min["Original Ask Amount"] / 10_00_000,
    color=[color_map[s] for s in result_min["Season Number"]]
)
ax[1].set_title("Lowest Original Ask")
ax[1].set_xlabel("Amount ($ Million)")
ax[1].bar_label(bars1, fmt="$%.2fM", padding=3)

# --- Legend ---
legend_patches = [
    Patch(color=color_map[s], label=f"Season {s}")
    for s in season_numbers
]

fig.legend(
    handles=legend_patches,
    title="Seasons",
    loc="lower center",
    ncol=6
)




plt.tight_layout(rect=[0, 0.08, 1, 0.92])
plt.show()
```

```python

sharks = [
    "Barbara Corcoran",
    "Mark Cuban",
    "Lori Greiner",
    "Robert Herjavec",
    "Daymond John",
    "Kevin O Leary",
]

colors = ['red', 'blue', 'green', 'orange', 'purple', 'cyan']

amount_cols = [
    'Barbara Corcoran Investment Amount',
    'Mark Cuban Investment Amount',
    'Lori Greiner Investment Amount',
    'Robert Herjavec Investment Amount',
    'Daymond John Investment Amount',
    'Kevin O Leary Investment Amount'
]

equity_cols = [
    'Barbara Corcoran Investment Equity',
    'Mark Cuban Investment Equity',
    'Lori Greiner Investment Equity',
    'Robert Herjavec Investment Equity',
    'Daymond John Investment Equity',
    'Kevin O Leary Investment Equity'
]

# -------------------------------
# Aggregate industry-wise
# -------------------------------
amount_data = df.groupby('Industry')[amount_cols].sum().fillna(0)
equity_data = df.groupby('Industry')[equity_cols].sum().fillna(0)

industry = amount_data.index

# -------------------------------
# Plot
# -------------------------------
fig, ax = plt.subplots(1, 2, figsize=(18, 12))
fig.suptitle(
    "Industry-wise Investment Amount & Equity (Stacked by Shark)",
    fontsize=14
)

# ---- LEFT Graph: Investment Amount ----
left_amount = np.zeros(len(industry))

for shark, col, color in zip(sharks, amount_cols, colors):
    ax[0].barh(
        industry,
        amount_data[col] / 10_00_000,   #  Lakhs
        left=left_amount,
        color=color
    )
    left_amount += amount_data[col] / 10_00_000

ax[0].set_title("Investment Amount ($ Million)")
ax[0].set_xlabel("Amount ($ Million)")
ax[0].set_ylabel("Industry")

# ---- RIGHT Graph: Investment Equity ----
left_equity = np.zeros(len(industry))

for shark, col, color in zip(sharks, equity_cols, colors):
    ax[1].barh(
        industry,
        equity_data[col],
        left=left_equity,
        color=color
    )
    left_equity += equity_data[col]

ax[1].set_title("Investment Equity (%)")
ax[1].set_xlabel("Equity (%)")

# -------------------------------
# Legend (shared)
# -------------------------------
legend_handles = [
    mpatches.Patch(color=color, label=shark)
    for shark, color in zip(sharks, colors)
]

fig.legend(
    handles=legend_handles,
    title="Sharks",
    loc='upper left',
    ncol=6,
    
)

plt.tight_layout(rect=[0, 0, 1, 0.93])
plt.show()
```

```python
for investor, (amount_col, equity_col) in investors.items():

    # ------------------- Amount Analysis -------------------
    invest_amt = (
        df[df[amount_col] > 0]
        .loc[:, ['Startup Name', amount_col]]
        .nlargest(10, amount_col)
    )

    industry_amt = (
        df[df[amount_col] > 0]
        .groupby('Industry')[amount_col]
        .sum()
        .sort_values()
    )

    # ------------------- Equity Analysis -------------------
    invest_eq = (
        df[df[equity_col] > 0]
        .loc[:, ['Startup Name', equity_col]]
        .nlargest(10, equity_col)
    )

    industry_eq = (
        df[df[equity_col] > 0]
        .groupby('Industry')[equity_col]
        .sum()
        .sort_values()
    )

    # ------------------- Plotting -------------------
    fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(22, 20))

    colors_amt = plt.cm.tab20.colors[:len(invest_amt)]
    colors_ind = plt.cm.tab20.colors[:len(industry_amt)]

    # Top 10 Investment Amount
    bars1 = ax1.bar(
        invest_amt['Startup Name'],
        invest_amt[amount_col] / 10_000,
        color=colors_amt
    )
    ax1.bar_label(bars1, fmt="$%.1f", padding=3)
    ax1.set_title(f"Top 10 {investor} Investments", fontsize=15)
    ax1.set_ylabel("Amount (in 10,000s)")
    ax1.set_xlabel("Startup Name")

    # Industry-wise Investment Amount
    bars2 = ax2.barh(
        industry_amt.index,
        industry_amt.values / 10_000,
        color=colors_ind
    )
    ax2.bar_label(bars2, fmt="$%.1f", padding=3)
    ax2.set_title("Industry-wise Investment")
    ax2.set_xlabel("Amount (in 10,000s)")

    # Top 10 Equity
    bars3 = ax3.bar(
        invest_eq['Startup Name'],
        invest_eq[equity_col],
        color=colors_amt
    )
    ax3.bar_label(bars3, fmt="%.1f", padding=3)
    ax3.set_title(f"Top 10 {investor} Equity Deals", fontsize=15)
    ax3.set_ylabel("Equity %")
    ax3.set_xlabel("Startup Name")


    # Industry-wise Equity
    bars4 = ax4.barh(
        industry_eq.index,
        industry_eq.values,
        color=colors_ind
    )
    ax4.bar_label(bars4, fmt="%.1f", padding=3)
    ax4.set_title("Industry-wise Equity")
    ax4.set_xlabel("Equity %")

    # Formatting
    plt.setp(ax1.get_xticklabels(), rotation=45, ha='right')
    plt.setp(ax3.get_xticklabels(), rotation=45, ha='right')
    plt.suptitle(investor, fontsize=18, fontweight='bold')
    plt.tight_layout()
    plt.show()
```

```python
investors = {
    "Barbara Corcoran": "Barbara Corcoran Investment Amount",
    "Mark Cuban": "Mark Cuban Investment Amount",
    "Lori Greiner": "Lori Greiner Investment Amount",
    "Robert Herjavec": "Robert Herjavec Investment Amount",
    "Daymond John": "Daymond John Investment Amount",
    "Kevin O Leary": "Kevin O Leary Investment Amount"
}
```

```python
for investor, amount_col in investors.items():

    # Season-wise total investment
    season_wise_amt = (
        df[df[amount_col] > 0]
        .groupby('Season Number')[amount_col]
        .sum()
        .astype(int)
    )

    # Season-wise deal count
    season_wise_count = (
        df[df[amount_col] > 0]
        .groupby('Season Number')[amount_col]
        .count()
    )

    colors = plt.cm.tab20.colors[:len(season_wise_amt)]

    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 8))

    # ---------------- Amount Plot ----------------
    bars1 = ax1.bar(
        season_wise_amt.index,
        season_wise_amt.values / 10_000,
        color=colors
    )
    ax1.bar_label(bars1, fmt="$%.1f", padding=3)
    ax1.set_xlabel("Season Number")
    ax1.set_ylabel("Investment Amount (in 10,000s)")
    ax1.set_title(f"Season-wise Investment Amount  {investor}")

    # ---------------- Count Plot ----------------
    bars2 = ax2.bar(
        season_wise_count.index,
        season_wise_count.values,
        color=colors
    )
    ax2.bar_label(bars2, fmt="%.0f", padding=3)
    ax2.set_xlabel("Season Number")
    ax2.set_ylabel("Number of Investments")
    ax2.set_title(f"Season-wise Investment Count  {investor}")
    
    plt.suptitle(investor, fontsize=18, fontweight='bold')
    plt.tight_layout()
    plt.show()
```
